```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('fifa_data.csv') # importing dataset
df.columns
```

```python
top_country = df['Nationality'].value_counts().idxmax()
top_country_count = df['Nationality'].value_counts().max()
print(f"Country with the most players: {top_country} ({top_country_count} players)")
```

**Country with the most players: England (1662 players)**

```python
top_5_countries = df['Nationality'].value_counts().head(5)

plt.figure(figsize=(7, 4))
top_5_countries.plot(kind='bar', color='skyblue')
plt.title('Top 5 Countries with Most Players')
plt.xlabel('Country')
plt.ylabel('Number of Players')
plt.show()
```

**Top 5 countries are England,Germany,Spain,Argentina,France**

```python
highest_salary_player = df.loc[df['Wage'].idxmax()]
print(f"Player with the highest salary: {highest_salary_player['Name']} with a salary of {highest_salary_player['Wage']}")
```

**Player with the highest salary: F. Marchetti with a salary of €9K**

```python
plt.figure(figsize=(7, 5))
plt.hist(df['Wage'], bins=20, color='green', edgecolor='black')
plt.title('Salary Range of Players')
plt.xlabel('Salary')
plt.ylabel('Number of Players')
plt.show()
```

```python
tallest_player = df.loc[df['Height'].idxmax()]
print(f"The tallest player: {tallest_player['Name']} with a height of {tallest_player['Height']} cm")
```

```python
top_club = df['Club'].value_counts().idxmax()
top_club_count = df['Club'].value_counts().max()
print(f"Club with the most players: {top_club} ({top_club_count} players)")
```

**Club with the most players: FC Barcelona (33 players)**

```python
preferred_foot = df['Preferred Foot'].value_counts()
print(f"Most preferred foot: {preferred_foot.idxmax()}")

plt.figure(figsize=(7, 5))
preferred_foot.plot(kind='bar', color='orange')
plt.title('Preferred Foot of Players')
plt.xlabel('Preferred Foot')
plt.ylabel('Number of Players')
plt.show()
```

**Most preffered foot is right foot**
